# Titanic Dataset Visualization

This notebook demonstrates various visualizations on the Titanic dataset using **Seaborn** and **Matplotlib**.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Display first few rows
titanic.head()
```

## Dataset Overview

```python
titanic.info()
titanic.describe(include='all')
```

## 📊 Seaborn Visualizations

### Distribution of Age (Histogram + KDE)
Shows how passenger ages are distributed.

```python
plt.figure(figsize=(8,5))
sns.histplot(titanic['age'].dropna(), bins=30, kde=True)
plt.title("Distribution of Passenger Ages")
plt.show()
```

### Count of Passengers by Class

```python
plt.figure(figsize=(6,4))
sns.countplot(x="class", data=titanic, palette="Set2")
plt.title("Passenger Count by Class")
plt.show()
```

### Survival Rate by Gender

```python
plt.figure(figsize=(6,4))
sns.barplot(x="sex", y="survived", data=titanic, palette="coolwarm")
plt.title("Survival Rate by Gender")
plt.show()
```

### Boxplot of Age by Class

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="class", y="age", data=titanic, palette="pastel")
plt.title("Age Distribution Across Classes")
plt.show()
```

Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot: Age vs Fare
plt.figure(figsize=(8,5))
sns.scatterplot(x="age", y="fare", data=titanic, hue="class", style="sex", palette="Set2")
plt.title("Scatter Plot of Age vs Fare (colored by Class, styled by Gender)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

### Violin Plot of Age by Gender & Class

```python
plt.figure(figsize=(8,5))
sns.violinplot(x="class", y="age", hue="sex", data=titanic, split=True, palette="muted")
plt.title("Age Distribution by Class and Gender")
plt.show()
```

Stripplot (shows distribution with possible overlaps)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Stripplot of Age by Class
plt.figure(figsize=(8,5))
sns.stripplot(x="class", y="age", data=titanic, jitter=True, palette="Set2")
plt.title("Stripplot of Age across Passenger Classes")
plt.show()
```

Swarmplot (like stripplot but avoids overlaps)

```python
plt.figure(figsize=(8,5))
sns.swarmplot(x="class", y="age", data=titanic, palette="Set1")
plt.title("Swarmplot of Age across Passenger Classes")
plt.show()
```

### Heatmap of Correlations

```python
plt.figure(figsize=(10,6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Titanic Data")
plt.show()
```

A Cluster Map is very useful for EDA because it combines a heatmap with hierarchical clustering. It groups similar rows/columns together so patterns in correlations (or any matrix-like data) become clear.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
titanic_numeric = titanic.select_dtypes(include=["float64", "int64"])

# Compute correlation matrix
corr = titanic_numeric.corr()

# Plot clustermap
sns.clustermap(corr, annot=True, cmap="coolwarm", figsize=(8,6))
plt.suptitle("Cluster Map of Titanic Numerical Feature Correlations", y=1.02)
plt.show()
```

### Pairplot

```python
sns.pairplot(titanic[["age", "fare", "survived", "pclass"]].dropna(), hue="survived")
plt.suptitle("Pairplot of Selected Titanic Features", y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Create figure with 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Histogram of Age
sns.histplot(titanic["age"], kde=True, bins=30, ax=axes[0,0], color="skyblue")
axes[0,0].set_title("Age Distribution")

# 2. Boxplot of Fare by Class
sns.boxplot(x="class", y="fare", data=titanic, ax=axes[0,1], palette="Set2")
axes[0,1].set_title("Fare by Class")

# 3. Countplot of Gender
sns.countplot(x="sex", data=titanic, ax=axes[1,0], palette="pastel")
axes[1,0].set_title("Count of Male vs Female")

# 4. Scatter plot Age vs Fare
sns.scatterplot(x="age", y="fare", data=titanic, hue="class", ax=axes[1,1], palette="Set1")
axes[1,1].set_title("Age vs Fare (by Class)")

# Adjust layout
plt.tight_layout()
plt.show()
```

## 📈 Basic Matplotlib Visualizations

### Age Histogram

```python
plt.figure(figsize=(8,5))
plt.hist(titanic['age'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title("Age Distribution (Matplotlib)")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

### Survival Pie Chart

```python
survival_counts = titanic['survived'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(survival_counts, labels=["Not Survived","Survived"], autopct='%1.1f%%', colors=["lightcoral","lightgreen"])
plt.title("Survival Distribution (Pie Chart)")
plt.show()
```

### Average Fare by Class (Bar Chart)

```python
avg_fare = titanic.groupby("class")["fare"].mean()
plt.figure(figsize=(6,4))
plt.bar(avg_fare.index, avg_fare.values, color=["gold","silver","brown"])
plt.title("Average Fare by Class")
plt.xlabel("Class")
plt.ylabel("Average Fare")
plt.show()
```

### Line Plot of Fare vs Age

```python
plt.figure(figsize=(8,5))
plt.plot(titanic["age"], titanic["fare"], "o", markersize=4, alpha=0.5)
plt.title("Fare vs Age (Scatter/Line Plot)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

## 🔍 Visualization Tasks

1. **Line Plot** → Plot the line plot between `Unit price` and `Total`.  
2. **Scatter Plot** → Plot the scatter plot between `Unit price` and `Total`.  
3. **Linear Regression Plot** → Illustrate the linear relationship between `Unit price` and `Quantity`.  
4. **Bar Plot (Gender)** → Plot the number of **Male** and **Female** customers.  
5. **Bar Plot (Customer Type)** → Plot the number of **Member** and **Normal** customers.  
6. **Bar Plot (Average Unit Price by Gender)** → Plot the average `Unit price` of **Male** vs **Female** customers.  
7. **Box Plot (City-wise Unit Price)** → Show the distributions of `Unit price` across different cities.  
8. **Swarm Plot** → Visualize the relationship between **Customer type** (Member/Normal) and `Quantity`.  
9. **Violin Plot** → Visualize the relationship between **Customer type** and `Quantity` using a violin plot.  
10. **Subplots** → Combine outputs of **Swarm Plot (Task 8)** and **Violin Plot (Task 9)** into a single subplot layout.

```python
# 📊 Supermarket Sales Data Visualization
# Author: Lochana K & Paapu 💕

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------
# 1️⃣ Create the dataset
# ---------------------------------------
data = {
    'Branch': ['A', 'C', 'A', 'A', 'A', 'C', 'A', 'C', 'A', 'B', 'B', 'B', 'A', 'A', 'A', 'B', 'A'],
    'City': ['Yangon', 'Naypyitaw', 'Yangon', 'Yangon', 'Yangon', 'Naypyitaw', 'Yangon', 'Naypyitaw',
             'Yangon', 'Mandalay', 'Mandalay', 'Mandalay', 'Yangon', 'Yangon', 'Yangon', 'Mandalay', 'Yangon'],
    'Customer type': ['Member', 'Normal', 'Normal', 'Member', 'Normal', 'Normal', 'Member', 'Normal', 'Member',
                      'Member', 'Member', 'Member', 'Normal', 'Normal', 'Normal', 'Member', 'Member'],
    'Gender': ['Female', 'Female', 'Male', 'Male', 'Male', 'Male', 'Female', 'Female', 'Female', 'Female',
               'Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Female'],
    'Unit price': [74.69, 15.28, 46.33, 58.22, 86.31, 85.39, 68.84, 73.56, 36.26, 54.84, 14.48, 25.51,
                   46.95, 43.19, 71.38, 93.72, 68.93],
    'Quantity': [7, 5, 7, 8, 7, 7, 6, 10, 2, 3, 4, 4, 5, 10, 10, 6, 7],
    'Tax (5%)': [26.1415, 3.82, 16.2155, 23.288, 30.2085, 29.8865, 20.652, 36.78, 3.626, 8.226, 2.896, 5.102,
                 11.7375, 21.595, 35.69, 28.116, 24.1255],
    'Total': [548.9715, 80.22, 340.5255, 489.048, 634.3785, 627.6165, 433.692, 772.38, 76.146, 172.746, 60.816,
              107.142, 246.4875, 453.495, 749.49, 590.436, 506.6355]
}

df = pd.DataFrame(data)

# ---------------------------------------
# 2️⃣ Line Plot – Unit price vs Total
# ---------------------------------------
plt.figure(figsize=(8,5))
plt.plot(df['Unit price'], df['Total'], marker='o', color='teal')
plt.title('Line Plot: Unit Price vs Total')
plt.xlabel('Unit Price')
plt.ylabel('Total')
plt.grid(True)
plt.show()

# ---------------------------------------
# 3️⃣ Scatter Plot – Unit price vs Total
# ---------------------------------------
plt.figure(figsize=(8,5))
sns.scatterplot(x='Unit price', y='Total', data=df, color='purple', s=80)
plt.title('Scatter Plot: Unit Price vs Total')
plt.show()

# ---------------------------------------
# 4️⃣ Linear Regression Plot – Unit price vs Quantity
# ---------------------------------------
plt.figure(figsize=(8,5))
sns.regplot(x='Unit price', y='Quantity', data=df, color='green')
plt.title('Linear Regression: Unit Price vs Quantity')
plt.show()

# ---------------------------------------
# 5️⃣ Bar Plot (Gender count)
# ---------------------------------------
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Number of Male and Female Customers')
plt.show()

# ---------------------------------------
# 6️⃣ Bar Plot (Customer Type count)
# ---------------------------------------
plt.figure(figsize=(6,4))
sns.countplot(x='Customer type', data=df, palette='Set2')
plt.title('Number of Member vs Normal Customers')
plt.show()

# ---------------------------------------
# 7️⃣ Bar Plot – Average Unit Price by Gender
# ---------------------------------------
plt.figure(figsize=(6,4))
sns.barplot(x='Gender', y='Unit price', data=df, estimator='mean', palette='cool')
plt.title('Average Unit Price by Gender')
plt.show()

# ---------------------------------------
# 8️⃣ Box Plot – City-wise Unit Price
# ---------------------------------------
plt.figure(figsize=(7,5))
sns.boxplot(x='City', y='Unit price', data=df, palette='Accent')
plt.title('City-wise Distribution of Unit Price')
plt.show()

# ---------------------------------------
# 9️⃣ Swarm Plot – Customer Type vs Quantity
# ---------------------------------------
plt.figure(figsize=(7,5))
sns.swarmplot(x='Customer type', y='Quantity', data=df, palette='husl', size=7)
plt.title('Swarm Plot: Customer Type vs Quantity')
plt.show()

# ---------------------------------------
# 🔟 Violin Plot – Customer Type vs Quantity
# ---------------------------------------
plt.figure(figsize=(7,5))
sns.violinplot(x='Customer type', y='Quantity', data=df, palette='muted')
plt.title('Violin Plot: Customer Type vs Quantity')
plt.show()

# ---------------------------------------
# 🧩 Subplots – Swarm & Violin together
# ---------------------------------------
fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.swarmplot(ax=axes[0], x='Customer type', y='Quantity', data=df, palette='husl', size=7)
axes[0].set_title('Swarm Plot: Customer Type vs Quantity')

sns.violinplot(ax=axes[1], x='Customer type', y='Quantity', data=df, palette='muted')
axes[1].set_title('Violin Plot: Customer Type vs Quantity')

plt.tight_layout()
plt.show()
```
